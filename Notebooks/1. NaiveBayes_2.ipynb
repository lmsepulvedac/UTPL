# Modelo Naive Bayes
0. Cargar las librerías de trabajo

```python
import pandas as pd
import numpy as np

from sklearn.naive_bayes import GaussianNB   #Colección de Reserva
from sklearn.metrics import confusion_matrix #La métrica de evaluación
```

1. Se hace la conexión con la Base de Datos

```python
nxl="/content/0. SolicitantesCrédito(USD).xlsx"
XDB=pd.read_excel(nxl,sheet_name=0)
XDB.head()       #Primeras lineas de visualización
XDB=XDB.dropna() #Eliminar elementos N/A
XDB=XDB[['Edad','Ingresos','Egresos','Monto','PreApr']]

#Cuales son las variables de entrada - la variable Referencia
XD=XDB[['Edad','Ingresos','Egresos','Monto']] #Entrada
yd=XDB[['PreApr']] #Valor Deseado - Valor de Referencia
```

2. Se procede con la implementación del modelo

```python
mnb=GaussianNB()   #Estructura General del Modelo
mnb.fit(XD,yd)     #Ajustamos el modelo
yp=mnb.predict(XD) #Probamos el funcionamiento del modelo
yd2=np.array(yd)   #Convertir el vector de referencia en array

mnb.theta_          #Estos son las medias de las variables
np.sqrt(mnb.var_)   #Estas son las desviaciones de las variables
```

3. Evaluamos el modelo - Matriz de Confusión

```python
cm=confusion_matrix(yd2,yp)

VP=cm[0,0] #Verdaderos Positivos
VN=cm[1,1] #Verdaderos Negativos
Exactitud=(VP+VN)/len(yd)
print("La exactitud del modelo es:",Exactitud)

FP=cm[1,0] #Falsos Positivos
FN=cm[0,1] #Falsos Negativos
TasaError=(FP+FN)/len(yd)
print("La tasa de error es:", TasaError)

#Sensibilidad - Que también el modelo clasifica los PreApr
Sensibilidad=VP/(VP+FP)
print("El nivel de sensibilidad es:",Sensibilidad)

#Especificidad - Que también el modelo clasifica los PreNeg
Especificidad=VN/(FN+VN)
print("El nivel de especificidad es:", Especificidad)

#Precisión
Precision=VP/(VP+VN)
print("La precision del modelo es:",Precision)
```
