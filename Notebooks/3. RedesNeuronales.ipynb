0. Cargar todas las librerias

```python
#!pip install tensorflow

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Librerias
from keras.models import Sequential #Es el cerebro por Cortes
from keras.layers import Dense      #Número de neuronas por Corte
```

1. Cargar los datos de la base de datos

```python
XDB=pd.read_excel('/content/0. SolicitantesCrédito(USD).xlsx',sheet_name=0)
XDB=XDB[['Edad','Ingresos','Egresos','Monto','Score']]
XD=XDB[['Edad','Ingresos','Egresos','Monto']] #Datos de Entrada
yd=XDB[['Score']]                             #Dato Salida - Referencia

#Normalizar las variables
XD=XD/np.max(XD,axis=0)   #axis=0, son los máximos por columna
yd=yd/np.max(yd,axis=0)
```

2. Diseñamos el modelo ADALINE (Adaptive Linear - solo una capa)

```python
NE=4
model=Sequential() #Estructura general del cerebro

model.add(Dense(1,activation='linear',use_bias=False,input_dim=NE)) #Cerebro por capas
model.compile(optimizer='adam',loss='mse')
history=model.fit(XD,yd,epochs=200,batch_size=1000)
yr=model.predict(XD)

#Procedemos con el despliegue de resultados
plt.figure()
plt.scatter(yd,yr)
model.layers[0].get_weights()
```
