0. Se procede con la carga de las librerías de trabajo

```python
import pandas as pd
import numpy as np

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Se procede con la carga de los datos

```python
nxl="/content/0. SolicitantesCrédito(USD).xlsx"
XDB=pd.read_excel(nxl,sheet_name=0)
XDB=XDB[['Edad','Ingresos','Egresos','Monto','PreApr']]
XDB=XDB.dropna()
XD=XDB[['Edad','Ingresos','Egresos','Monto']]
yd=XDB[['PreApr']]
```

2. Se procede con la construcción del modelo

```python
mnb=GaussianNB()
mnb.fit(XD,yd)
mnb.theta_  #Esto es para encontrar los centroides asociados a cada variable
np.sqrt(mnb.var_)    #Esto representa la varianza
```

3. Se procede con la evaluación del modelo

```python
ydd=np.array(yd)
ypred=mnb.predict(XD)
confusion_matrix(ydd,ypred)
```

4. Si se quiere evaluar la preaprobación de una persona con la siguientes características

```python
Xs=np.array([[34,1500,1000,3000]])
yps=mnb.predict(Xs)
print("La decisión frente al crédito es:",yps)
```
